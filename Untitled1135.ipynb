```python
import pandas as pd
import os

# Load dataset
csv_path = r"C:\Users\sagni\Downloads\House Pricing\archive (1)\socal2.csv"
img_folder = r"C:\Users\sagni\Downloads\House Pricing\archive (1)\socal2\socal_pics"
output_html = r"C:\Users\sagni\Downloads\House Pricing\house_search.html"

df = pd.read_csv(csv_path)

# Build HTML content
html_content = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>House Search</title>
    <style>
        body { font-family: Arial, sans-serif; margin: 20px; background: #f5f5f5; }
        h1 { color: #333; }
        form { margin-bottom: 20px; }
        input[type="text"], input[type="number"] {
            padding: 10px; width: 200px; margin-right: 10px; border-radius: 5px; border: 1px solid #ccc;
        }
        button {
            padding: 10px 15px; background-color: #4CAF50; color: white; border: none; border-radius: 5px;
            cursor: pointer;
        }
        button:hover { background-color: #45a049; }
        .house {
            display: inline-block; margin: 10px; border: 1px solid #ccc; background: #fff;
            border-radius: 5px; overflow: hidden; width: 300px; box-shadow: 0 0 5px rgba(0,0,0,0.2);
        }
        .house img { width: 100%; height: auto; }
        .details { padding: 10px; }
    </style>
</head>
<body>
    <h1>Find Houses by Location and Price</h1>
    <form id="searchForm">
        <input type="text" id="location" placeholder="Enter Location (e.g., Brawley)" required>
        <input type="number" id="minPrice" placeholder="Min Price" required>
        <input type="number" id="maxPrice" placeholder="Max Price" required>
        <button type="button" onclick="searchHouses()">Search</button>
    </form>
    <div id="results"></div>

    <script>
        const houses = """ + df.to_json(orient="records") + """;

        function searchHouses() {
            const location = document.getElementById('location').value.toLowerCase();
            const minPrice = parseInt(document.getElementById('minPrice').value);
            const maxPrice = parseInt(document.getElementById('maxPrice').value);
            const resultsDiv = document.getElementById('results');
            resultsDiv.innerHTML = '';

            const filtered = houses.filter(h => 
                h.citi.toLowerCase().includes(location) &&
                h.price >= minPrice &&
                h.price <= maxPrice
            );

            if (filtered.length === 0) {
                resultsDiv.innerHTML = '<p>No houses found for your criteria.</p>';
                return;
            }

            filtered.slice(0, 12).forEach(house => {
                const imgPath = './socal2/socal_pics/' + house.image_id + '.jpg';
                const houseDiv = document.createElement('div');
                houseDiv.className = 'house';
                houseDiv.innerHTML = `
                    <img src="${imgPath}" alt="House Image">
                    <div class="details">
                        <strong>${house.street}</strong><br>
                        â‚¹${house.price.toLocaleString()}<br>
                        ${house.bed} bed, ${house.bath} bath, ${house.sqft} sqft
                    </div>
                `;
                resultsDiv.appendChild(houseDiv);
            });
        }
    </script>
</body>
</html>
"""

# Write HTML to file
with open(output_html, "w", encoding="utf-8") as f:
    f.write(html_content)

print(f"HTML file created at: {output_html}")
```
